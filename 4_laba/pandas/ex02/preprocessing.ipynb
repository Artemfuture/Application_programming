## Exercise 02. Preprocessing

1. Считываю CSV-файл, создав ID столбец с индексами.

```python
import pandas as pd

df = pd.read_csv("../data/auto.csv", index_col="ID")
df.head()
```

2. Подсчитываю количество наблюдений, используя метод count().

```python
df.count()
```

3. Удаляю дубликаты, оставив только следующие столбцы: `CarNumber`, `Make_n_Model` и `Fines`

```python
df = df.drop_duplicates(subset=["CarNumber", "Make_n_model", "Fines"], keep="last")
df.count()
```

4. Работа с отсутствующими значениями

```python
df.isna().sum()
```

```python
df = df.dropna(axis=1, thresh= len(df) - 500) 
df.isna().sum()
```

Заменяем отсутствующие значения предудущим значением 

```python
df["Refund"] = df["Refund"].ffill()
df.isna().sum() 
```

Заменяем `Fines` средними значениями этого столбца

```python
fines_mean = df["Fines"].mean()
df["Fines"] = df["Fines"].fillna(fines_mean)
df.isna().sum()
df.head()
```

5. Разделяю и анализирую марку и модель.

```python
df["Make"] = df["Make_n_model"].apply(lambda x: str(x).split()[0])
df["Model"] = df["Make_n_model"].apply(lambda x: " ".join(str(x).split()[1:]))
```

```python
df = df.drop(columns=["Make_n_model"])
df.head()
```

```python
df.to_json("../data/auto.json", orient="records", indent=4)
```
