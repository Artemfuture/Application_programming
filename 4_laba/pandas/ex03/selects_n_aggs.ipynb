## Exercise 03. Selects and aggregations

1. Загрузите файл JSON, который вы создали в предыдущем упражнении, в датафрейм.

```python
import pandas as pd

df = pd.read_json("../data/auto.json")
df = df.set_index("CarNumber")
df.head()

```

2.1 Отображаю строки, в которых сумма штрафов превышает 2100.

```python
df[df["Fines"] > 2100]
```

2.2 Отображаю строки, в которых штрафы превышают 2100, а возврат средств равен 2.

```python
df[(df["Fines"] > 2100) & (df["Refund"] == 2)]
```

2.3 Отображаю строки, в которых модели взяты из списка: [’Focus’, ’Corolla’].


```python
df[df["Model"].isin(["Focus", "Corolla"])]
```

2.4 Отображаю строки, в которых номер автомобиля совпадает с номером из списка: [`Y7689C197RUS`, `92928M178RUS`, `7788KT197RUS`, `H115YO163RUS`, `X758HY197RUS`].

```python
numbers = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df.loc[df.index.isin(numbers)]
```

3. Выполняю агрегацию с помощью `make` и `model`

Медианные штрафы сгруппированные по `Make`

```python
df.groupby("Make")["Fines"].median()
```

Медианные штрафы, сгруппированные по `Make` и `Model`

```python
df.groupby(["Make", "Model"])["Fines"].median()
```

Количество штрафов, сгруппированных по `Make` и `Model`

```python
df.groupby(["Make", "Model"])["Fines"].count()

```

Минимальные и максимальные штрафы, сгруппированные по `Make` и `Model`

```python
df.groupby(["Make", "Model"])["Fines"].agg(["min", "max"])
```

Стандартное отклонение штрафов, сгруппированных по `Make` и `Model`

```python
df.groupby(["Make", "Model"])["Fines"].std()

```

4. Выполняю агрегацию по номерам автомобилей

Номера автомобилей, сгруппированные по количеству штрафов в порядке убывания

```python
top_by_count = df.groupby(df.index)["Fines"].count().sort_values(ascending=False)
top_by_count
```

Чаще всего нарушал:

```python
top1_by_count = top_by_count.index[0]
df[df.index == top1_by_count]
```

Номера автомобилей, сгруппированные по сумме штрафов в порядке убывания

```python
top_by_sum = df.groupby(df.index)["Fines"].sum().sort_values(ascending=False)
top_by_sum
```

```python
top1_by_sum = top_by_sum.index[0]
df[df.index == top1_by_sum]
```

Связаны ли какие-либо автомобильные номера с разными моделями? (Да)

```python
car = df.groupby(df.index)['Model'].nunique()
car[car > 1]
```
