## Exercise 01. Basic operations

1. Создайте фрейм данных под названием `views` с двумя столбцами: `datetime` и `user`, прочитав `feed-views.log`

```python
import pandas as pd

views = pd.read_csv(
    "../data/feed-views.log",
    sep="\t",
    names=["datetime", "user"]
)

views.head()
```

2. Создаём новую колонку **daytime**

```python
views["datetime"] = pd.to_datetime(views["datetime"], format="%Y-%m-%d %H:%M:%S.%f")
```

```python
views["year"] = views["datetime"].dt.year
views["month"] = views["datetime"].dt.month
views["day"] = views["datetime"].dt.day
views["hour"] = views["datetime"].dt.hour
views["minute"] = views["datetime"].dt.minute
views["second"] = views["datetime"].dt.second
```

```python
labels = ["night", "early morning", "morning", "afternoon", "early evening", "evening"]
bins = [0, 4, 7, 11, 17, 20, 24]

views["daytime"] = pd.cut(views["hour"], bins=bins, labels=labels, right= False)
```

```python
views = views.set_index("user")
views.head()
```

3. Подсчитываем количество элементов в нашем датафрейме

```python
count = views.count()
count
```

Подсчитываем колличество элементов в views["daytime"]

```python
views["daytime"].value_counts()
```

4. Сортируем значения в вашем датафрейме по часам, минутам и секундам в порядке возрастания одновременно, а не по отдельности.

```python
sorted_views = views.sort_values(by=["hour", "minute", "second"])
sorted_views.head()
```

5. Рассчитываем минимальное и максимальное значения для часов и режим для дневных категорий.

```python
hour_min = views["hour"].min()
hour_max = views["hour"].max()
daytime_mode = views["daytime"].mode().iloc[0]
(hour_min, hour_max, daytime_mode)
```

```python
night_data = views[views["daytime"] == "night"]
max_night_hour = night_data["hour"].max()
user = night_data[night_data["hour"] == max_night_hour].index[0]
(max_night_hour, user)
```

```python
morning_data = views[views["daytime"] == "morning"]
min_morning_hour = morning_data["hour"].min()

user = morning_data[morning_data["hour"] == min_morning_hour].index[0]
(min_morning_hour, user)
```

```python
hour_mode = views["hour"].mode()[0]

(hour_mode, daytime_mode)
```

6. Отображаю три самых ранних и самых поздних часа дня и соответствующие им имена пользователей с помощью nsmallest() и nlargest().

```python
earliest = views.nsmallest(3, "hour")[["hour", 'daytime']]
latest = views.nlargest(3, "hour")[["hour", 'daytime']]

earliest, latest
```
